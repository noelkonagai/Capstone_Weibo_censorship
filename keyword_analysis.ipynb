```python
import pandas as pd
import numpy as np
from pypinyin import pinyin, lazy_pinyin, Style
```

```python
keywords = pd.read_csv('./data/keyword_query/keywords_all.txt',sep="\n")
```

```python
keywords.columns = ['Link']
```

```python
keywords['Keywords'] = [l.split('s.weibo.com/weibo/')[-1] for l in keywords['Link']]
keywords['Pinyin'] = [lazy_pinyin(w) for w in keywords['Keywords']]
```

```python
trimmed = keywords[~(keywords.Keywords.str.contains('=|%|\?'))]
```

```python
trimmed.head(10)
```

```python
politics_keys = ['习','江','蛤','共','党','匪','法轮','邪教','台湾','香港','天安门','64','六四','达赖','薄熙来','周永康','周恩来','毛泽东','小平']
politics_dict = {key:[] for key in politics_keys}
for w in trimmed['Keywords']:
    for k in politics_keys:
        if k in w:
            politics_dict[k].append(w)
```

```python
sex_keys = ['性','色情','淫','激情','做爱','AV']
sex_dict = {key:[] for key in sex_keys}
for w in trimmed['Keywords']:
    for k in sex_keys:
        if k in w:
            sex_dict[k].append(w)
```
